# Edge-Based Segmentation

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load an image from a file
image = cv2.imread('download.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print('Error: Could not open or read the image')
    exit()

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Canny edge detector
edges = cv2.Canny(blurred, threshold1=30, threshold2=100)

# Show the original image and the edges using cv2_imshow
cv2_imshow(image)
cv2_imshow(edges)
```

# Threshold-Based Segmentation

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load an image from a file (grayscale)
image = cv2.imread('download.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print('Error: Could not open or read the image')
    exit()

# Set a threshold value (you can adjust this as needed)
threshold_value = 128

# Apply binary thresholding
_, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Show the original image and the binary segmented image using cv2_imshow
cv2_imshow(image)
cv2_imshow(binary_image)
```

# Region-Based Segmentation

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load an image from a file (grayscale)
image = cv2.imread('download.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print('Error: Could not open or read the image')
    exit()

# Threshold the image to create a binary image
threshold_value = 128
_, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

# Find connected components in the binary image
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_image, connectivity=8)

# Create a color image to display the segmented regions
segmented_image = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)

# Loop through each connected component and assign a random color
for label in range(1, num_labels):
    color = np.random.randint(0, 256, size=3)
    segmented_image[labels == label] = color

# Show the original image and the segmented image using cv2_imshow
cv2_imshow(image)
cv2_imshow(segmented_image)
```
