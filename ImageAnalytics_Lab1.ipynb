```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt



# Function to apply contrast stretching
def apply_contrast_stretching(image):
    min_intensity = image.min()
    max_intensity = image.max()
    stretched_image = ((image - min_intensity) / (max_intensity - min_intensity) * 255).astype('uint8')
    return stretched_image

# Function to apply histogram equalization
def apply_histogram_equalization(image):
    # Convert to grayscale if not already
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    equalized_image = cv2.equalizeHist(image)
    return equalized_image


# Function to adjust brightness and contrast
def adjust_brightness_contrast(image, alpha, beta):
    adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted_image


# Function to apply sharpening
def apply_sharpening(image):
    kernel = np.array([[-1, -1, -1],
                       [-1, 9, -1],
                       [-1, -1, -1]])
    sharpened_image = cv2.filter2D(image, -1, kernel)
    return sharpened_image



# Function to upload an image
def upload_image():
    uploaded = files.upload()
    if len(uploaded) == 0:
        return None
    file_name = list(uploaded.keys())[0]
    image = cv2.imread(file_name)
    return image, file_name


# Upload an image
uploaded_image, image_filename = upload_image()


if uploaded_image is not None:
    # Apply image enhancement techniques
    enhanced_image1 = apply_contrast_stretching(uploaded_image)
    enhanced_image2 = apply_histogram_equalization(uploaded_image)
    enhanced_image4 = adjust_brightness_contrast(uploaded_image, alpha=1.5, beta=10)
    enhanced_image5 = apply_sharpening(uploaded_image)

    # Display the original and enhanced images
    plt.figure(figsize=(14, 10))

    plt.subplot(2, 3, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(uploaded_image, cv2.COLOR_BGR2RGB))

    plt.subplot(2, 3, 2)
    plt.title("Contrast Stretching")
    plt.imshow(enhanced_image1, cmap='gray')

    plt.subplot(2, 3, 3)
    plt.title("Histogram Equalization")
    plt.imshow(enhanced_image2, cmap='gray')


    plt.subplot(2, 3, 4)
    plt.title("Brightness and Contrast Adjustment")
    plt.imshow(cv2.cvtColor(enhanced_image4, cv2.COLOR_BGR2RGB))

    plt.subplot(2, 3, 5)
    plt.title("Sharpened Image")
    plt.imshow(cv2.cvtColor(enhanced_image5, cv2.COLOR_BGR2RGB))

    plt.tight_layout()
    plt.show()
else:
    print("No image uploaded.")
```

```python
import numpy as np
import skimage.io as skio
from skimage import img_as_ubyte, img_as_float
guinanpicard = skio.imread('mri.jpg')
for channel, color in enumerate('rgb'):
    channel_values = guinanpicard[:,:,channel]
    plt.step(np.arange(256),
    np.bincount(channel_values.flatten(), minlength=256)*1.0/channel_values.size, c=color)
    plt.xlim(0, 255)
    plt.axvline(np.percentile(channel_values, 95), c=color)
    plt.xlabel('channel value')
    plt.ylabel('fraction of pixels');
```

We could see from above that the maximum value of each of the color channels are at 255, however this also means that applying a direct white patch to our image would not change the results. So rather than using the maximum value, we will use the 95th percentile of their values for the white patch:

```python
guinanpicard_gw = ((guinanpicard* (guinanpicard.mean() / guinanpicard.mean(axis=(0, 1)))).clip(0, 255).astype(int))
skio.imshow(guinanpicard_gw);
```

```python
guinanpicard_wp = img_as_ubyte((guinanpicard * 1.0 / np.percentile(guinanpicard, 95, axis=(0, 1))).clip(0, 1))
skio.imshow(guinanpicard_wp)
```
