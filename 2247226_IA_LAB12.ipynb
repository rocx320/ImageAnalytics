```python
import cv2
import os
import shutil

# Path to the folder containing frames
folder_path = '/content/sample_data/FrameFolder'

# Create folders for each emotion
emotions = ["Happy", "Sad"]
for emotion in emotions:
    emotion_folder = os.path.join(folder_path, emotion)
    if not os.path.exists(emotion_folder):
        os.makedirs(emotion_folder)

# Load the Haar Cascade Classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Process each frame in the folder
for filename in os.listdir(folder_path):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)

        if image is None:
            print(f"Error reading image: {filename}")
            continue

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        if len(faces) == 0:
            print(f"No faces detected in image: {filename}")
            continue

        for (x, y, w, h) in faces:
            # Move the image to the corresponding emotion folder
            emotion_label = "Happy"  # Placeholder, replace with actual emotion classification
            destination_folder = os.path.join(folder_path, emotion_label)
            destination_path = os.path.join(destination_folder, filename)
            cv2.imwrite(destination_path, image)

            print(f"Image {filename} moved to {destination_folder}")

# Zip the emotion folders
shutil.make_archive("/content/sample_data/EmotionFrames", 'zip', folder_path)

print("Images have been moved to emotion folders and zipped.")
```
