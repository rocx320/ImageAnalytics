```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as compare_ssim
import matplotlib.pyplot as plt

# Load the original and noisy images
original_image = cv2.imread('lena4.jpg')
noisy_image = cv2.imread('lena6.jpg') #noisy


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(original_image)

plt.subplot(1, 2, 2)
plt.title('Noisy Image ')
plt.imshow(noisy_image)
```

Apply the quality measures to check the image quality

```python

noisy_image = cv2.resize(noisy_image, (original_image.shape[1], original_image.shape[0]))

# Calculate RMSE
rmse = np.sqrt(((original_image - noisy_image) ** 2).mean())

# Calculate PSNR
psnr = cv2.PSNR(original_image, noisy_image)

# Calculate SSIM (convert to grayscale)
ssim = compare_ssim(cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY), cv2.cvtColor(noisy_image, cv2.COLOR_BGR2GRAY))

# Create a bar graph to compare the metrics
metrics = ['RMSE', 'PSNR', 'SSIM']
values = [rmse, psnr, ssim]

plt.bar(metrics, values)
plt.xlabel('Metric')
plt.ylabel('Value')
plt.title('Comparison of Denoising Metrics')
plt.show()
```
