```python
import cv2
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error, peak_signal_noise_ratio
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate RMSE, PSNR, and SSIM
def calculate_metrics(image1, image2):
    rmse = np.sqrt(mean_squared_error(image1, image2))
    psnr = peak_signal_noise_ratio(image1, image2)
    ssim_score = ssim(image1, image2, multichannel=True)
    return rmse, psnr, ssim_score

# Load the original image
original_image = cv2.imread('/content/img5.jpeg')

if original_image is None:
    raise Exception("The original image could not be loaded.")

# Create a copy of the original image and add noise to it
noisy_image = original_image.copy()
noise = np.random.normal(0, 25, noisy_image.shape).astype(np.uint8)
noisy_image = cv2.add(noisy_image, noise)

# Calculate metrics for the noisy image compared to the original image
rmse, psnr, ssim_score = calculate_metrics(original_image, noisy_image)

# Plot the metrics as a bar graph and display the images
metrics = ['RMSE', 'PSNR', 'SSIM']
values = [rmse, psnr, ssim_score]

# Create a single row with the images and the graph
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Display the original image
axes[0].imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')

# Display the noisy image
axes[1].imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
axes[1].set_title("Noisy Image")
axes[1].axis('off')

# Plot the metrics as a bar graph
axes[2].bar(metrics, values, color=['blue', 'green', 'orange'])
axes[2].set_ylabel('Metric Value')
axes[2].set_title("Comparison")

plt.tight_layout()
plt.show()

# Print the calculated values
for metric, value in zip(metrics, values):
    print(f"{metric}: {value:.2f}")
```
